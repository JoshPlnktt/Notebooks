### Exploring Wealth by Industry

```python
SELECT
	industry,
	SUM(CAST(networth AS INT)) AS total_wealth
FROM billionaires.billionaires
GROUP BY industry
ORDER BY total_wealth DESC
```

### Next, I will explore the difference in total billionaires by gender

```python
SELECT
	gender,
	COUNT(gender) AS total_billionaires
FROM billionaires.billionaires
GROUP BY gender
```
